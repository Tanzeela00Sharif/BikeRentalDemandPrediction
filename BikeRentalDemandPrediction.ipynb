# Results

1️⃣ Decision Tree Regression
R² Score: 0.8905 → This means your Decision Tree model explains ~89% of the variation in bike rental counts in your test data.

That’s a very strong fit — predictions are close to actual values most of the time.

2️⃣ Linear Regression
R² Score: 0.3593 → This means the model explains ~36% of the variation.

That’s relatively low, so the model misses many patterns in the data.

3️⃣ Which is better?
Based only on R², Decision Tree Regression is performing much better on your dataset.

Why?

Bike rentals probably depend on non-linear relationships (e.g., sudden changes in demand during rush hours, weather effects, holidays).

Linear Regression can only model straight-line relationships, so it struggles with complex patterns.

Decision Trees can split data into segments and capture such non-linear patterns better.
